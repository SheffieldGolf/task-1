## 1. N-gram: (length 1 2 3) tf.idf, tf

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer
from collections import Counter

# N-gram length: 1-3. using tf construct vocab.
count_vec = CountVectorizer(ngram_range=(1, 3), max_features=8000)
tf = count_vec.fit_transform(tweets_df['preprocessed_text_with_stemming'].values.astype('U')).toarray()
vocab = count_vec.get_feature_names()

# td.idf
tfidf_vec = TfidfVectorizer(ngram_range=(1, 3), min_df=1, vocabulary=vocab)
tfidf= tfidf_vec.fit_transform(tweets_df['preprocessed_text_with_stemming'].values.astype('U')).toarray()
```

```python
print(vocab[:20])
```

```python
# reconstruct dataframe.
data_tf = pd.DataFrame(tf, columns=vocab)
data_tf.head(5)
```

```python
data_tfidf = pd.DataFrame(tfidf, columns=vocab)
data_tfidf.head(5)
```

```python
data_tf.to_csv("data_stem_tf_features.csv")
data_tfidf.to_csv("data_stem_tfidf_features.csv")
```

## 2. Word-embedding: Word-2-Vec

```python
from gensim import corpora
from gensim.models import Word2Vec

texts = list()
for d in tweets_df["preprocessed_text_no_stemming"].values.astype('U'):
    d1 = d.split(" ")
    texts.append(d1)

w2v_model = Word2Vec(texts, size=200, window=5, workers=4)
w2v_model.wv.save_word2vec_format('word_embedding_output') # save model and word vector.
```

## 3. part-of-speech:

```python
nltk.download('averaged_perceptron_tagger')
nltk.download('universal_tagset')
nltk.pos_tag(texts[0], tagset='universal')
```

```python
# add pos to original texts.
pos_list = list()
pos_text = list()
for d in texts:
    pos = nltk.pos_tag(d, tagset='universal')
    pos_list.append([p[0]+"("+p[1]+")" for p in pos])
for t in pos_list:
    pos_text.append(str(" ").join(t))
```

```python
pos_text
```

```python
# N-gram length: 1-3. using tf construct vocab.
count_vec = CountVectorizer(ngram_range=(1, 3), max_features=8000, token_pattern=r"^[A-Za-z]+[\S][A-Za-z]+[\S]", lowercase=False)
tf_pos = count_vec.fit_transform(pos_text).toarray()
vocab_pos = count_vec.get_feature_names()

# td.idf
tfidf_vec = TfidfVectorizer(ngram_range=(1, 3), min_df=1, vocabulary=vocab_pos)
tfidf_pos = tfidf_vec.fit_transform(pos_text).toarray()
```

```python
# reconstruct dataframe.
data_tf_pos = pd.DataFrame(tf_pos, columns=vocab_pos)
data_tf_pos.to_csv("data_stem_tf_pos.csv")
```

```python
data_tfidf_pos = pd.DataFrame(tfidf_pos, columns=vocab_pos)
data_tfidf_pos.to_csv("data_stem_tfidf_pos.csv")
```
